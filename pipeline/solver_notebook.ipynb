```python
import numpy as np
from solver import Grid, TimeGrid, PhysicalParams, BoundaryParams, Masks, RadiatorParams, HeatEquationSolver2D

grid = Grid(nx=80, ny=50, hx=0.05)
time = TimeGrid(dt=0.01, t_end=50.0)
phys = PhysicalParams(alpha=0.02)
bc = BoundaryParams(kind="dirichlet", dirichlet_value=273.15)  # na zewnątrz 0°C (prosty test)

u0 = np.ones(grid.shape) * 285.15  # 12°C start

# maska grzejnika: mały prostokąt przy dolnej ścianie
rad = np.zeros(grid.shape, dtype=bool)
rad[45:48, 10:25] = True

masks = Masks(radiator=rad)
radiator = RadiatorParams(enabled=True, setpoint=293.15, coeff=0.02)  # 20°C

solver = HeatEquationSolver2D(grid, time, phys, bc, masks=masks, radiator=radiator)
out = solver.simulate(u0, store_every=50)

# szybkie sanity-checki:
print("start mean:", out["U"][0].mean() - 273.15)
print("end mean:", out["U"][-1].mean() - 273.15)
```

```python
import numpy as np
from solver import Grid, TimeGrid, PhysicalParams, BoundaryParams, Masks, RadiatorParams, HeatEquationSolver2D
from metrics import mean_temp_C, std_temp_C

grid = Grid(nx=60, ny=40, hx=0.05)
time = TimeGrid(dt=0.02, t_end=30.0)
phys = PhysicalParams(alpha=0.02)
bc = BoundaryParams(kind="dirichlet", dirichlet_value=273.15)  # 0°C na brzegu

u0 = np.ones(grid.shape) * 285.15  # 12°C

rad = np.zeros(grid.shape, dtype=bool)
rad[35:38, 5:20] = True

masks = Masks(radiator=rad)
radiator = RadiatorParams(enabled=True, setpoint=293.15, coeff=0.02)

solver = HeatEquationSolver2D(grid, time, phys, bc, masks=masks, radiator=radiator)
out = solver.simulate(u0, store_every=25)

print("Mean start/end (°C):", mean_temp_C(out["U"][0]), mean_temp_C(out["U"][-1]))
print("Std  start/end (°C):", std_temp_C(out["U"][0]),  std_temp_C(out["U"][-1]))
print("Psi end:", out["Psi"][-1])
```
