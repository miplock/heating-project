## 1.1. Czy grzejnik musi być pod oknem?

### Pytanie badawcze
W tym eksperymencie sprawdzamy, czy umieszczenie grzejnika pod oknem poprawia rozkład temperatury w pomieszczeniu oraz czy wpływa na komfort cieplny względem alternatywnych lokalizacji grzejnika.

### Intuicje i oczekiwane wyniki
- P1: [intuicja / oczekiwany trend]

### Schemat układu pomieszczenia

Rozważamy jednopokojową domenę prostokątną
$$
\Omega = (0,L_x)\times(0,L_y),
$$
którą w całości utożsamiamy z pokojem $R_1=\Omega$.

**Okno.** Okno umieszczamy na górnej ścianie ($y=L_y$) jako odcinek o szerokości $w$ wycentrowany w osi $x$:
$$
W_{\text{win}}=\{(x,y)\in\partial\Omega:\ y=L_y,\ x\in[x_c-\tfrac{w}{2},\,x_c+\tfrac{w}{2}]\},\qquad x_c=\tfrac{L_x}{2}.
$$
Pozostała część brzegu stanowi ściany zewnętrzne:
$$
W_{\text{wall}}=\partial\Omega\setminus W_{\text{win}}.
$$

**Grzejnik (źródło ciepła).** Modelujemy grzejnik jako prostokątny podobszar $R_{\text{rad}}\subset\Omega$ o szerokości $d$ i wysokości $h_{\text{rad}}$.

- **Scenariusz A (grzejnik „pod oknem”)**:
$$
R_{\text{rad}}^{(A)}=[x_c-\tfrac{d}{2},\,x_c+\tfrac{d}{2}]\times[y_0,\,y_0+h_{\text{rad}}].
$$

- **Scenariusz B (grzejnik na przeciwległej ścianie)**:
$$
R_{\text{rad}}^{(B)}=[x_c-\tfrac{d}{2},\,x_c+\tfrac{d}{2}]\times[L_y-y_0-h_{\text{rad}},\,L_y-y_0].
$$

W eksperymencie parametryzujemy położenie grzejnika odległością $r$ od okna (wzdłuż osi $y$) oraz jego rozmiarem $d$, a jako miary komfortu analizujemy średnią temperaturę $\mu_u(t)$ oraz niejednorodność rozkładu poprzez odchylenie standardowe $\sigma_u(t)$.

## 1.2. Czy wyłączać grzejniki przed wyjściem z domu?

### Pytanie badawcze
W tym eksperymencie porównujemy strategie ogrzewania podczas nieobecności domowników: ciągłe grzanie (np. na niższym nastawie) vs. wyłączenie ogrzewania i późniejsze dogrzanie po powrocie.

### Intuicje i oczekiwane wyniki
- P2: [intuicja / oczekiwany trend]

### Schemat układu pomieszczeń

Modelujemy mieszkanie jako domenę prostokątną
$$
\Omega=(0,L_x)\times(0,L_y),
$$
podzieloną na rozłączne pokoje $R_i$ (maski pomieszczeń). Przyjmujemy podział na 5 stref:
$$
\Omega=\bigcup_{i=1}^{5} R_i,\qquad R_i\cap R_j=\emptyset\ \ (i\neq j).
$$

**Podział na pomieszczenia (przykładowa konfiguracja).** Dla ustalonych $L_x,L_y$ definiujemy:
- sypialnia:
$$
R_{\text{syp}}=(0,\tfrac{L_x}{2})\times(\tfrac{L_y}{2},L_y),
$$
- salon:
$$
R_{\text{sal}}=(\tfrac{L_x}{2},L_x)\times(\tfrac{L_y}{2},L_y),
$$
- łazienka:
$$
R_{\text{laz}}=(0,\tfrac{3}{10}L_x)\times(0,\tfrac{L_y}{2}),
$$
- przedpokój:
$$
R_{\text{prz}}=(\tfrac{3}{10}L_x,\tfrac{L_x}{2})\times(0,\tfrac{L_y}{2}),
$$
- kuchnia:
$$
R_{\text{kuc}}=(\tfrac{L_x}{2},L_x)\times(0,\tfrac{L_y}{2}).
$$

**Okna i ściany zewnętrzne.** Okna umieszczamy na wybranych fragmentach brzegu $\partial\Omega$ (np. jedno okno w salonie na $y=L_y$ oraz jedno w sypialni na $x=0$). Zbiór okien oznaczamy przez
$$
W_{\text{win}}\subset \partial\Omega,
$$
a pozostały brzeg traktujemy jako ściany zewnętrzne:
$$
W_{\text{ext}}=\partial\Omega\setminus W_{\text{win}}.
$$
Ściany wewnętrzne odpowiadają granicom między pokojami (fragmenty $\partial R_i\cap \partial R_j$).

**Grzejniki (źródła ciepła).** W wybranych pokojach definiujemy podobszary grzejników
$$
R_{\text{rad},i}\subset R_i,
$$
np. $R_{\text{rad},\text{sal}}\subset R_{\text{sal}}$ i $R_{\text{rad},\text{syp}}\subset R_{\text{syp}}$ (pozostałe pokoje mogą mieć $f_i\equiv 0$). Grzejniki sterowane są termostatem w każdym pokoju, tj. źródło w $R_i$ aktywuje się, gdy średnia temperatura w pokoju spada poniżej nastawy $S_i(t)$.

**Horyzont czasowy i strategie.** Ustalamy stały czas eksperymentu $T_{\text{total}}$ obejmujący wyjście i powrót. Porównujemy:
- strategię ciągłego grzania (stałe nastawy $S_i(t)\equiv S_{\text{base}}$),
- strategię „wychładzanie + dogrzewanie” (nastawy zależne od czasu):
$$
S_i(t)=
\begin{cases}
S_{\text{low}}, & t\in[0,t_{\text{out}}],\\
S_{\text{high}}, & t\in(t_{\text{out}},T_{\text{total}}].
\end{cases}
$$
Jako kryteria porównania przyjmujemy zużytą energię $\Psi(T_{\text{total}})$ oraz czas odzyskania komfortu (np. osiągnięcie zadanej temperatury średniej w salonie i sypialni).

## Stałe fizyczne wykorzystane w modelu

### Stała gazowa powietrza
Indywidualna stała gazowa powietrza oznaczana jest symbolem $r$ i występuje w równaniu stanu gazu doskonałego. Określa zależność pomiędzy ciśnieniem, temperaturą oraz gęstością powietrza.  
Dla suchego powietrza przyjęto wartość:
$$
r = 287\ \mathrm{J/(kg\cdot K)}.
$$
Wartość ta wynika z ilorazu uniwersalnej stałej gazowej i masy molowej powietrza i jest standardowo stosowana w modelach termodynamicznych.  
Źródło:  
https://www.engineeringtoolbox.com/air-properties-d_156.html

---

### Ciepło właściwe powietrza
Ciepło właściwe powietrza przy stałym ciśnieniu oznaczane jest symbolem $c$ i określa ilość energii potrzebnej do podniesienia temperatury jednostki masy powietrza o 1 K.  
Dla powietrza w temperaturze pokojowej przyjęto wartość:
$$
c = 1005\ \mathrm{J/(kg\cdot K)}.
$$
Wartość ta odpowiada zakresowi temperatur typowych dla pomieszczeń mieszkalnych i jest powszechnie wykorzystywana w obliczeniach inżynierskich.  
Źródło:  
https://www.engineeringtoolbox.com/air-specific-heat-capacity-d_705.html

---

### Ciśnienie
Ciśnienie powietrza oznaczane jest symbolem $p$ i w modelu traktowane jest jako stałe, odpowiadające warunkom atmosferycznym.  
Przyjęto wartość:
$$
p = 101325\ \mathrm{Pa}.
$$
Jest to standardowe ciśnienie atmosferyczne na poziomie morza, często stosowane jako wartość referencyjna w modelach fizycznych.  
Źródło:  
https://www.engineeringtoolbox.com/standard-atmosphere-d_604.html

---

### Przewodnictwo cieplne powietrza
Przewodnictwo cieplne powietrza oznaczane jest symbolem $\alpha$ i w modelu pełni rolę efektywnego współczynnika dyfuzji ciepła w powietrzu. Parametr ten opisuje szybkość rozprzestrzeniania się temperatury w ośrodku i występuje w równaniu dyfuzji ciepła.  
Dla powietrza w temperaturze pokojowej przyjęto wartość:
$$
\alpha = 2.1 \times 10^{-5}\ \mathrm{m^2/s}.
$$
Jest to wartość odpowiadająca molekularnej dyfuzyjności cieplnej powietrza; w rzeczywistych warunkach pomieszczeń może ona być traktowana jako wartość efektywna, uwzględniająca słabą konwekcję.  
Źródło:  
https://www.engineeringtoolbox.com/air-thermal-diffusivity-d_201.html
