# Exp1: wczytanie gotowych wykresow

Notebook wczytuje zapisane wykresy z plikow `.pkl` i tylko je wyswietla.


```python
from pathlib import Path
import json
import subprocess
import sys
import pickle
```

### parametry eksperymentalne

```python
PROJECT_ROOT = Path.cwd().parents[0]
EXPERIMENTAL_PATH = (
    PROJECT_ROOT / "data" / "exp1_params_experimental.json"
)

experimental_params = {
    "room_width": 4.0,
    "room_height": 3.0,
    "hx": 0.05,
    "t_end": 1.0,
    "dt": 0.5,
    "alpha": 2.1e-05,
    "u_out_c": 10.0,
    "u0_c": 16.0,
    "setpoint_c": 24.0,
    "rad_coeff": 0.2,
    "window_width": 1.2,
    "rad_width": 1.0,
    "rad_height": 0.3,
    "rad_center_x": None,
    "bc_kind": "dirichlet",
}

EXPERIMENTAL_PATH.parent.mkdir(parents=True, exist_ok=True)
EXPERIMENTAL_PATH.write_text(
    json.dumps(experimental_params, indent=2, ensure_ascii=True)
    + "\n"
)
EXPERIMENTAL_PATH
```

### uruchamianie eksperymentu

```python
import subprocess
import sys

RUN_EXP1 = True
USE_EXPERIMENTAL = True
SHOW_PROGRESS = True

runner_path = PROJECT_ROOT / "pipeline" / "exp1" / "runner.py"
cmd = [sys.executable, str(runner_path)]
if USE_EXPERIMENTAL:
    cmd += ["--use-experimental"]
if SHOW_PROGRESS:
    cmd += ["--progress"]
if RUN_EXP1:
    subprocess.run(cmd, check=True, cwd=PROJECT_ROOT)
```

### wizualizacja

```python
import subprocess
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[0]
script_path = PROJECT_ROOT / "pipeline" / "exp1" / "visualize.py"
MAPS_PATH = PROJECT_ROOT / "data" / "exp1_maps.npz"
SHOW_PROGRESS = False
cmd = [
    sys.executable,
    str(script_path),
    "--maps-input",
    str(MAPS_PATH),
]
if SHOW_PROGRESS:
    cmd += ["--progress"]
subprocess.run(cmd, check=True, cwd=PROJECT_ROOT)
```

### wyświetlanie wykresów

```python
FIG_DIR = PROJECT_ROOT / "data" / "exp1_figures"

fig_paths = sorted(FIG_DIR.glob("*.pkl"))
if not fig_paths:
    raise FileNotFoundError(
        f"Brak plikow .pkl w katalogu: {FIG_DIR}"
    )

for path in fig_paths:
    with path.open("rb") as handle:
        fig = pickle.load(handle)
    fig.show()
```
