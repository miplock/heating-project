dla danego pokoju mamy:

$$
\frac{\partial u}{\partial t} = \alpha \Delta u + f_m(x,y,u)
$$

gdzie

$$
f_m(x,y, u) =
\begin{cases}
u \cdot \beta, 
& (x,y) \in R_m,\ \dfrac{1}{|R_m|}\displaystyle\int_{R_m} u(x,y,t)\,d\omega \le S_i, \\[1em]
0, & \text{w przeciwnym razie}.
\end{cases}
$$
i $\beta = \dfrac{P \cdot r}{p \cdot |R_i| \cdot c}$

funkcję $f$ możemy również przedstawić następująco:

$$
f(\textbf x, u(\cdot, t)) = \beta \cdot u(\cdot, t)\cdot \textbf{1}\left(\textbf x \in \mathcal R_m\right) \cdot \textbf{1}\left(\frac{1}{|R_m|}\int_{R_m}u(t, \textbf x))d\omega < S_i\right)
$$

oznaczmy
$$
f_1(\textbf{x}):=\textbf{1}(\textbf{x} \in \mathcal R_m)
$$
czyli (dyskretyzując):
$$
f_1(\textbf{x}_i)=\textbf{1}(\textbf{x}_i \in \mathcal R_m)
$$
teraz $f_2$:
$$
f_2(u(\cdot, t)):=\textbf{1}\left(\frac{1}{|R_m|}\int_{R_m}u(t, \textbf x))d\omega < S_i\right)
$$
dyskretyzując
$$
f_2(u^n):=\textbf{1}\left(\frac{1}{|R_m|}\sum_{\Delta R_m}u^n_i \Delta R_m < S_i\right)=\textbf{1}\left(\frac{1}{|R_m|}\sum_{i}u^n_i \cdot h^2 < S_i\right)
$$
tutaj tego $h^2$ nie jestem do konca pewny, ale to bedzie cos w tym stylu

tu będą takie trapezy 2D, w każdym razie $f_2$ jest zależny tylko od $u$ (i od dyskretyzacji siatki ale to niewazne)
